### Convert datatypes

```python
data['Date'] = pd.to_datetime(data['Date'], errors= 'coerce')
data['Updated On'] = pd.to_datetime(data['Updated On'], errors='coerce')
print(data.info())
```

```python
data
```

```python
data['Year'] = data['Date'].dt.year
crime_count_per_year = data['Year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_count_per_year.index, y=crime_count_per_year.values)
plt.title('Crime Count per Year')
plt.xlabel('Year')
plt.ylabel('Crime Count')
plt.show()
```

```python
crime_counts_per_type = data['Primary Type'].value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(y=crime_counts_per_type.index, x=crime_counts_per_type.values)
plt.title('Crime Count per Primary Type')
plt.xlabel('Crime Count')
plt.ylabel('Primary Type')
plt.show()
```

```python
# Identify non-numeric columns
non_numeric_columns = data.select_dtypes(include=['object']).columns
print(non_numeric_columns)
```

```python
# Drop unnecessary columns and inspect remaining non-numeric columns
columns_to_drop = ['ID', 'Case Number', 'Date', 'Block', 'Description', 'Updated On', 'Location']
crime_data = data.drop(columns=columns_to_drop)

# Identify remaining non-numeric columns
non_numeric_columns = crime_data.select_dtypes(include=['object']).columns
print(non_numeric_columns)

# Apply one-hot encoding to the non-numeric columns
crime_data = pd.get_dummies(crime_data, columns=non_numeric_columns, drop_first=True)
```

```python
# Verify all columns are numeric
print(crime_data.info())
```

```python
from sklearn.model_selection import train_test_split

# Define features and target variable
features = crime_data.drop('Year', axis=1)
target = crime_data['Year']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# Example: Predict crime rates for a new dataset (replace `new_data` with actual data)
new_data = X_test.iloc[:5]  # Using a subset of the test set as an example
predictions = model.predict(new_data)

print('Predicted Crime Rates:', predictions)
```
